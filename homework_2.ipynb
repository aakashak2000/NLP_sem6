```
import pandas as pd
```

```
df = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz", lines=True)
df.head()
```

```
df['reviewText_count'] = df.reviewText
df['summary_count'] = df.summary
```

```
def word_occurences(string):
    count=dict()
    words=string.split()
    for word in words:
      if word in count:
        count[word]+=1
      else:
        count[word]=1
    return count

df['reviewText_wordcount']=df['reviewText'].apply(word_occurences)
df['summary_wordcount']=df['summary'].apply(word_occurences)
```

```
df.head()
```

```
def letters(string):
    vowels = 0
    consonants = 0
    words = string.count(' ') + 1
    for i in range(0, len(string)):  
            ch = string[i]  
            ch = ch.lower()
            if ((ch >= 'a' and ch <= 'z') or (ch >= 'A' and ch <= 'Z')):
             if (ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u'): 
                vowels += 1
             elif (ch != 'a' or ch != 'e' or ch != 'i' or ch != 'o' or ch != 'u'): 
                consonants += 1
          
    count = {'vowels': vowels, 'consonants': consonants, 'words': words}
    return count
```

```
df["reviewText_letter_count"] = df.reviewText.apply(letters)
df["summary_letter_count"] = df.summary.apply(letters)
```

```
def BG(string):
    string = re.sub('[^a-zA-Z]',' ', string) #removes punctuations
    list1 = re.findall(r"\bw+[Gg]\b", string) #finding letters ending with g/G
    list2 = re.findall(r"\b[Bb]\w+", string) #finding letters starting with b/B
    return (list(set().union(list1, list2)))
```

```
BG('bass and treble of a song') #example
```

```
df["reviewText_BG"] = df.reviewText.apply(BG)
df["summary_BG"] = df.summary.apply(BG)
```

```
def not_BG(string):
    string = re.sub('[^a-zA-Z]',' ', string) #removes punctuations
    list1 = re.findall(r"\b(?![bB])\w+", string) #finding words not ending with g/G and words not starting with b/B
    list2 = re.findall(r"\b\w+(?<![gG])\b", string) #finding words not ending with g/G and words not starting with b/B
    return (list(set(list1).intersection(list2)))
```

```
not_BG('bass and treble of a song') #example
```

```
df["reviewText_not_BG"] = df.reviewText.apply(not_BG)
df["summary_not_BG"] = df.summary.apply(not_BG)
```
