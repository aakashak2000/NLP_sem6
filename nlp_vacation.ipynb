```
import numpy as np
import pandas as pd
import nltk
import re
from sklearn.feature_extraction.text import CountVectorizer
```

```
df_twitter=pd.read_csv('https://raw.githubusercontent.com/zfz/twitter_corpus/master/full-corpus.csv',error_bad_lines=False)
df_amazon=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
df_twitter=df_twitter.head(1000)
```

```
df_twitter.head()
```

```
from nltk import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english')) 
nltk.download('punkt')
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def punc(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
def stop_w(text): #Tokenizing text and removing the stop words
  text=text.lower()
  tokens=word_tokenize(text)
  nostop=[words for words in tokens if not words in stop_words]
  return nostop
```

```
from string import digits 
def no_digits(list):
    no_dig = str.maketrans('', '', digits) 
    list = [i.translate(no_dig) for i in list] 
    return list
```

```
def list_to_str(list):
  str=" "
  return (str.join(list))
```

```
df_twitter['TweetText_Clean']=df_twitter['TweetText'].apply(punc)
df_twitter['TweetText_Clean']=df_twitter['TweetText_Clean'].apply(stop_w)
df_twitter['TweetText_Clean']=df_twitter['TweetText_Clean'].apply(no_digits)
df_twitter['TweetText_Clean']=df_twitter['TweetText_Clean'].apply(list_to_str)
```

```
df_twitter['TweetText_Clean'][0]
```

```
from sklearn.model_selection import train_test_split as tts
count_vec=CountVectorizer()
```

```
X=count_vec.fit_transform(df_twitter['TweetText_Clean'])
X_train,X_test,y_train,y_test=tts(X,df_twitter['Sentiment'],test_size=0.3,random_state=20)
```

```
print(count_vec.get_feature_names())
```

```
X_train=pd.DataFrame(X_train.toarray(),columns=count_vec.get_feature_names())
X_test=pd.DataFrame(X_test.toarray(),columns=count_vec.get_feature_names())
```

```
X_train
```

```
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```
rfc=RandomForestClassifier()
```

```
params_rf={'n_estimators':[5,10,50,100],'max_depth':[10,15,20],'min_samples_leaf':[2,4],'max_features':['sqrt']}
```

```
model_rf=GridSearchCV(estimator=rfc,param_grid=params_rf,cv=5)
```

```
model_rf.fit(X_train,y_train)
```

```
y_pred_rf=model_rf.predict(X_test)
print(classification_report(y_pred_rf,y_test))
```

```
import lightgbm as lgbm
```

```
params_lgbm={'objective':'multiclass','num_class':3,'max_depth':10,'feature_fraction':0.8,}
```

```
from sklearn.preprocessing import LabelEncoder
```

```
le=LabelEncoder()
train_data=lgbm.Dataset(X_train,le.fit_transform(y_train))
```

```
model_lgbm=lgbm.train(params_lgbm,train_data)
```

```
y_pred_lgbm=model_lgbm.predict(X_test)
```

```
best_pred=[np.argmax(val) for val in y_pred_lgbm]
```

```
print(classification_report(le.inverse_transform(best_pred),y_test))
```

```
def thr_1(array):
  arr=[]
  for i in range(len(array)):
    if max(array[i])>0.6:
      arr.append(np.argmax(array[i]))
    else:
      arr.append(2)
  return arr
```

```
def thr_2(array):
  arr=[]
  for i in range(len(array)):
    if max(array[i])>0.7:
      arr.append(np.argmax(array[i]))
    else:
      arr.append(2)
  return arr
```

```
def thr_3(array):
  arr=[]
  for i in range(len(array)):
    if max(array[i])>0.8:
      arr.append(np.argmax(array[i]))
    else:
      arr.append(2)
  return arr
```

```
y_predprob_rf=model_rf.predict_proba(X_test)
print('Classification report with default threshold: \n',classification_report(model_rf.predict(X_test),y_test))
print('\n Classification report with 0.6 threshold: \n',classification_report(le.inverse_transform(thr_1(y_predprob_rf)),y_test))
print('\n Classification report with 0.7 threshold: \n',classification_report(le.inverse_transform(thr_2(y_predprob_rf)),y_test))
print('\n Classification report with 0.8 threshold: \n',classification_report(le.inverse_transform(thr_3(y_predprob_rf)),y_test))
```

```
y_pred_lgbm=model_lgbm.predict(X_test)
best_pred=[np.argmax(val) for val in y_pred_lgbm]
print('Classification report with default threshold: \n',classification_report(le.inverse_transform(best_pred),y_test))
print('\n Classification report with 0.6 threshold: \n',classification_report(le.inverse_transform(thr_1(y_pred_lgbm)),y_test))
print('\n Classification report with 0.7 threshold: \n',classification_report(le.inverse_transform(thr_2(y_pred_lgbm)),y_test))
print('\n Classification report with 0.8 threshold: \n',classification_report(le.inverse_transform(thr_3(y_pred_lgbm)),y_test))
```
